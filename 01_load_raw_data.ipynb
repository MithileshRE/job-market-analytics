```python
# NOTEBOOK 01: Load Raw Data
# Purpose: Read CSV from OneLake and create raw staging table
# ============================================================

from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType

# Step 1: Define Schema
schema = StructType([
    StructField("job_id", IntegerType(), True),
    StructField("job_title", StringType(), True),
    StructField("company_name", StringType(), True),
    StructField("location", StringType(), True),
    StructField("salary_min", IntegerType(), True),
    StructField("salary_max", IntegerType(), True),
    StructField("job_description", StringType(), True),
    StructField("posted_date", StringType(), True)
])

# Step 2: Read CSV from OneLake Files folder
df_raw = spark.read.schema(schema).csv("abfss://HashTech@onelake.dfs.fabric.microsoft.com/job_market_analytics_lh.Lakehouse/Files/sample_job_postings.csv", header=True)

# Step 3: Display preview
print("Raw Data Preview:")
df_raw.show(5)
print(f"Total Records: {df_raw.count()}")

# Step 4: Create table in lakehouse
df_raw.write.mode("overwrite").format("delta").saveAsTable("stg_job_postings_raw")

print("✅ Table 'stg_job_postings_raw' created successfully!")

# Step 5: Verify table creation
spark.sql("SELECT COUNT(*) as total_records FROM stg_job_postings_raw").show()
```
